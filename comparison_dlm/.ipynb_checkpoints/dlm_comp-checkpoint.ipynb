## 状態空間モデルを各ライブラリで試す
### 試したライブラリ
- Statsmodels
- PyStan


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# データセットの読み込み
dateparse = lambda dates: pd.datetime.strptime(dates, '%Y')
df = pd.read_csv('data/nile.csv', index_col='year', date_parser=dateparse)
#df.plot()
ts = df['volume']
plt.plot(ts)
```

```python
# Statsmodelsの状態空間モデル（ローカルレベルモデル）
import statsmodels.api as sm

model = sm.tsa.UnobservedComponents(ts, 'local level')
result = model.fit()
#display(result.summary())
result.plot_components(figsize=(8, 8))
plt.show()
```
