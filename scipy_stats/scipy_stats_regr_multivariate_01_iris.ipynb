```python
%matplotlib inline
import pandas as pd
from scipy import stats

df = pd.read_csv('examples/iris.csv')
df.tail(30)
```

### とりあえずpairplotで可視化してみる

```python
import seaborn as sb
sb.pairplot(df, hue='name')
```

### petal_lengthがsepal_widthに与える影響で、nameによる違いがあるかを検定する

```python
import statsmodels.formula.api as smf

model = smf.ols('sepal_width ~ name + petal_length', df)
result = model.fit()
print(result.summary())
```

### 分散分析による仮説検定
分散分析: analysis of variance (ANOVA)  
OLS Regression Results の coef の違いが有意かどうかを検定  

```python
print(result.f_test([0, 1, -1, 0]))
```
